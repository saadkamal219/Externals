```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
x = update_dataframe.drop(['G3'], axis=1)
```

```python
statement = pd.DataFrame(selectkbest.scores_, index=x.columns, columns=['Score'])
selected_statement = statement.nlargest(32, 'Score')
selected_statement
```

**Algorithm Implementation**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.6, random_state=3)
```

Attempt 1

```python
linearModel = LinearRegression()
linearModel.fit(x, y)
```

```python
y_pred = linearModel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np

print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred))
print("Mean squared Error:     ", mean_squared_error(y_test, y_pred))
print("Root Mean squared Error:", np.sqrt(mean_squared_error(y_test, y_pred)))

result = r2_score(y_test, y_pred)

print("Accuracy:               ", result*100)
```

Attempt 2


```python
from sklearn import tree

dicission_tree = tree.DecisionTreeClassifier()

dicission_tree.fit(x_train, y_train)
```

```python
y_pred_tree = dicission_tree.predict(x_test)
```

```python
print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred_tree))
print("Mean squared Error:     ", mean_squared_error(y_test, y_pred_tree))
print("Root Mean squared Error:", np.sqrt(mean_squared_error(y_test, y_pred_tree)))

result = r2_score(y_test, y_pred_tree)

print("Accuracy:               ", result*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(x_train,y_train)

y_pred_knn = knn.predict(x_test)

print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred_knn))
print("Mean squared Error:     ", mean_squared_error(y_test, y_pred_knn))
print("Root Mean squared Error:", np.sqrt(mean_squared_error(y_test, y_pred_knn)))

result = r2_score(y_test, y_pred_knn)

print("Accuracy:               ", result*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=21)
rf.fit(x_train,y_train)

y_pred_rf = rf.predict(x_test)

print("Mean Absolute Error:    ", mean_absolute_error(y_test, y_pred_rf))
print("Mean squared Error:     ", mean_squared_error(y_test, y_pred_rf))
print("Root Mean squared Error:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))

result = r2_score(y_test, y_pred_rf)

print("Accuracy:               ", result*100)
```

# After using multiple algorithms, we can see that Linear Regression comes with the highest accuracy 87.888% ~ 88%. So it is the final algorithm for my project
