**Null value handling**  ✅

```python
#fillna tech

meanAge = dataframe['Age'].mean()
medianAge = dataframe['Age'].median()
modeAge = dataframe['Age'].mode()

update_dataframe3 = dataframe.copy()

update_dataframe3['Age'].fillna(meanAge, inplace=True)
```

```python
#dropna tech

update_dataframe = dataframe.dropna()
```

```python
#drop tech

update_dataframe2 = dataframe.drop(['Age'], axis=1)
update_dataframe2
```

```python
update_dataframe.info()
```

**Encoding**

```python
#onehot encoding

onehot_dataframe = pd.get_dummies(update_dataframe, columns=['Gender', 'Geography'])
```

```python
onehot_dataframe
```

```python
#label encoding

from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

label_dataframe = update_dataframe.copy()

label_dataframe['Gender'] = labelencoder.fit_transform(label_dataframe['Gender'])
label_dataframe['Geography'] = labelencoder.fit_transform(label_dataframe['Geography'])
```

```python
label_dataframe.sample(15)
```

```python
#ordinal_encoding
update_dataframe['Geography'].unique()
```

```python
update_dataframe['Gender'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinalencoder = OrdinalEncoder(categories=[['France', 'Spain', 'Germany'], ['Male', 'Female']])

ordinal_dataframe = update_dataframe.copy()

ordinal_dataframe[['Geography', 'Gender']] = ordinalencoder.fit_transform(ordinal_dataframe[['Geography', 'Gender']])
```

```python
ordinal_dataframe.head()
```

```python
x = ordinal_dataframe.drop(['Surname','Exited'], axis=1)
```

```python
y = ordinal_dataframe['Exited']
```

```python
y
```

**Feature Selection**

```python
ordinal_dataframe.describe()
```

```python
ordinal_dataframe.corr()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
fit_features = SelectKBest(score_func = f_classif)
fit_features.fit(x,y)
```

```python
statement = pd.DataFrame(fit_features.scores_, index=x.columns, columns=['Relation'])
statement
```

```python
selected_statement = statement.nlargest(5, 'Relation')
selected_statement
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(x,y)
```

```python
statement_by_tree_classifier = pd.DataFrame(model.feature_importances_, index=x.columns, columns=['Relation'])
statement_by_tree_classifier
```

```python
selected_statement_by_tree_classifier = statement_by_tree_classifier.nlargest(5,'Relation')
selected_statement_by_tree_classifier
```

Decission Tree

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=5)
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score

dicission_tree = tree.DecisionTreeClassifier()

dicission_tree.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(25,20))
display = tree.plot_tree(dicission_tree)
```

```python
y_test_predict = dicission_tree.predict(x_test)
score = accuracy_score(y_test_predict, y_test)
print("Accuracy by using dicission tree classifier = ", score*100, "%")
```
